```python
corr_to_target = (
    data.corr(numeric_only=True)["heart_attack"]
      .sort_values(ascending=False)
)
print(corr_to_target)

sns.heatmap(data.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
